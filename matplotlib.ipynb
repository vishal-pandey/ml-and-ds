## X vs Y

```python
x = [1,2,3,4,5,6]
y = [1,2,2,3,3,4]
plt.plot(x, y)
plt.show()
```

## Plot types

```python
plt.plot(x, y, 'ro')
plt.show()
```

```python
# 'ro' = red dot
# r = red, b = blue , g = green,
# o = dot, - = dash, -- = double dashed, s = squqre, ^ = triangle
```

## Categorical Plotting

```python
names = ['delhi', 'goa', 'mumbai']
values = [1, 10, 50]
```

```python
plt.figure(figsize=(15, 5))


plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values, color='r', linewidths=10)
plt.subplot(133)
plt.plot(names, values, 'g--', linewidth=3.0)
plt.show()
```

## Working with text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
```

```python
# print(np.random.randn(10000))
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel("Smarts")
plt.ylabel("Probablity")
plt.title("Histogram of IQ")
plt.text(60, 0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```
