```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Getting data

```python
dataset = pd.read_csv('../datasets/housing.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, :-2].values
```

```python
y = dataset.iloc[:, -2:-1].values
```

## Discovery and visualisation

```python
dataset.keys()
```

```python
X[3]
```

```python
for i in range(0,8):
    plt.plot(X[:, i], y, 'go')
    plt.title(r"Median house value vs "+dataset.keys()[i])
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
X_new = X[:, 7]
```

```python
X_new = X_new.reshape(-1,1)
```

```python
# y.reshape(,-1)
```

```python
lin_reg.fit(X_new, y)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
predicted = lin_reg.predict(X_new)
```

```python
plt.plot(X[:, 7], y, 'go')
plt.plot(X_new, predicted, 'r--')
plt.show()
```

```python
lin_reg.score(X_new, y)
```
