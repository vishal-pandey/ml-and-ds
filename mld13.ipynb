# Titanic Dataset (Decision Tree Classifier)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data collection

```python
dataset = pd.read_csv("datasets/titanic/train.csv")
```

## Data Visualisation

```python
dataset
```

## Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
X = dataset.iloc[:, :].values
```

```python
y = dataset.iloc[:, [1]].values
```

```python
y
```

```python
imp = SimpleImputer()
lab = LabelEncoder()
```

```python
y = pd.DataFrame(y)
```

```python
y[0].value_counts()
```

```python
y = y[0].fillna('0')
```

```python
y
```

```python
# y = lab.fit_transform(y)
```

```python
# y
```

```python
# output_classes = lab.classes_
```

```python
X = X[:, [0, 1, 2, 4, 5, 6, 9, 11]]
```

```python
len(X[:, 3])
```

```python
X[:, 3] = lab.fit_transform(X[:, 3])
```

```python
lab.classes_
```

```python
temp = pd.DataFrame(X[:, 7])
```

```python
temp[0].value_counts()
```

```python
temp[0] = temp[0].fillna('S')
```

```python
# temp[0]
```

```python
# X[:, 7]
```

```python
X[:, 7] = lab.fit_transform(temp)
```

```python
X = imp.fit_transform(X)
```

```python
X
```

## Training

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_clf = DecisionTreeClassifier(max_depth=13)
```

```python
tree_clf
```

```python
tree_clf.fit(X, y)
```

## Model Visualisation

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(
    tree_clf,
    out_file="titanic_tree.dot",
    class_names = ['0', '1'],
    rounded = True,
    filled = True
)
```

## Testing (On Training Set)

```python
tree_clf.predict(X) == y
```

```python
tree_clf.score(X, y)
```
