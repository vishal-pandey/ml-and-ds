```python
# Urls to scrap the reviews from
urls = [
    "https://www.gartner.com/reviews/market/meeting-solutions-web-conferencing/vendor/att/product/att-connect?pageNum=1",
    "https://www.gartner.com/reviews/market/meeting-solutions-web-conferencing/vendor/att/product/att-connect?pageNum=2",
    "https://www.gartner.com/reviews/market/meeting-solutions-web-conferencing/vendor/att/product/att-connect?pageNum=3",
    "https://www.gartner.com/reviews/market/meeting-solutions-web-conferencing/vendor/att/product/att-connect?pageNum=4",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=1",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=2",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=3",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=4",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=5",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=6",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=7",
    "https://www.gartner.com/reviews/market/network-services/vendor/att?pageNum=8",
    
]
```

```python
# List for storing scrapped data
reviews = []
```

```python
# Opening a chrome window for loading the pages
# Path is of the webdriver for chrome, 
# if not present in your computer download from this link [https://sites.google.com/a/chromium.org/chromedriver/home]

browser = webdriver.Chrome("/chromedriver")
```

```python
# Visiting every links from the urls list and gathering various data about reviews
for link in urls:
    res = browser.get(link)
    time.sleep(10)
    page_content = browser.page_source
    soup = BeautifulSoup(page_content)
    cards = soup.find_all("div", {'class':'reviewSnippetCard'})
    
    
    for card in cards:
        title = card.find('section', {'class':'snippetBody'}).find('h3', {'itemprop': 'name'}).text
        review = card.find('p', {'class':'snippetSummary'}).text
        date_p = card.find('p', {'itemprop': 'datePublished'}).text
        rating = int(card.find('span', {'class': 'stars-icon-star-on'})['style'][6:-2])/20
        reviewer_role1 = card.find('dl', {'class':'hidden-xs'}).find_all('dd')[0].text
        company_size = reviewer_role = card.find('dl', {'class':'hidden-xs'}).find_all('dd')[1].text
        designation = card.find('h5', {'class': 'snippetJobAndIndustry'}).find_all('section')[0].text[3:]
        industury = card.find('h5', {'class': 'snippetJobAndIndustry'}).find_all('section')[1].text
        
        reviews.append([title,review,date_p,rating,reviewer_role1,company_size,designation,industury])
```

# 3.) Data Preprocessing

```python
# Clubbing the title and review column of the dataset to make the feature matrix X and y as labels (output)
X = dataset.iloc[:, 0].values +" "+ dataset.iloc[:, 1].values
y = dataset.iloc[:, -1].values
```

```python
# Importing the libraries for natural language processing and regular expressions
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# X
```

## Cleaning the data

```python
# Here we removed all the characters except alphabets in english and then comverted them to lower case
processed_list = []
for i in range(len(X)):
    review = re.sub('[^a-zA-Z]', ' ', X[i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(token) for token in review if not token in stopwords.words('english')]
    review = " ".join(review)
    processed_list.append(review)
```

```python
processed_list[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=3000)
```

```python
X = cv.fit_transform(processed_list).toarray()
```

```python
# X
```

```python
print(cv.get_feature_names())
```

```python
# Splitting the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# 4.) Select and train ML algo

#### Naive bayes is selected as it performs best on textual data

```python
from sklearn.naive_bayes import GaussianNB
```

```python
n_b = GaussianNB()
```

```python
n_b.fit(X_train, y_train)
```

```python
# Score on training set
n_b.score(X_train, y_train)
```

```python
# Score on testing set
n_b.score(X_test, y_test)
```

```python
# Score on whole data
n_b.score(X, y)
```
